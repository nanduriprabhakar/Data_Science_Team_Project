```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as plt2
from statsmodels.tsa.arima_model import ARIMA
%matplotlib inline
```

```python
raw_data_f1=pd.read_csv('File_1_copy.csv',index_col='time')
raw_data_f2=pd.read_csv('File_2_copy.csv',index_col='time')
raw_data_f3=pd.read_csv('File_3_copy.csv',index_col='time')
```

```python
input_and_output_f1=pd.DataFrame(columns=['leaf1','leaf2','leaf3','leaf4','leaf5','leaf6','leaf7','leaf8','spine1','spine2','spine3','spine4'], index=raw_data_f1.index)
input_and_output_f2=pd.DataFrame(columns=['leaf1','leaf2','leaf3','leaf4','leaf5','leaf6','leaf7','leaf8','spine1','spine2','spine3','spine4'], index=raw_data_f2.index)
input_and_output_f3=pd.DataFrame(columns=['leaf1','leaf2','leaf3','leaf4','leaf5','leaf6','leaf7','leaf8','spine1','spine2','spine3','spine4'], index=raw_data_f3.index)
```

```python
for i in input_and_output_f1.columns:
    input_and_output_f1[i]=raw_data_f1[i+' input packet rate'] + raw_data_f1[i+' output packet rate']
for i in input_and_output_f2.columns:
    input_and_output_f2[i]=raw_data_f2[i+' input packet rate'] + raw_data_f2[i+' output packet rate']
for i in input_and_output_f3.columns:
    input_and_output_f3[i]=raw_data_f3[i+' input packet rate'] + raw_data_f3[i+' output packet rate']
```

```python
input_and_output_f1_timechange=input_and_output_f1
input_and_output_f1_timechange.index=pd.to_datetime(input_and_output_f1_timechange.index,unit='ns')
input_and_output_f2_timechange=input_and_output_f2
input_and_output_f2_timechange.index=pd.to_datetime(input_and_output_f2_timechange.index,unit='ns')
input_and_output_f3_timechange=input_and_output_f3
input_and_output_f3_timechange.index=pd.to_datetime(input_and_output_f3_timechange.index,unit='ns')
```

```python
pd.tools.plotting.autocorrelation_plot(input_and_output_f2_timechange['leaf1'].dropna())
```

```python
model = ARIMA(input_and_output_f2_timechange['leaf1'].dropna(), order=(1,1,0))
```

```python
model_fit=model.fit(disp=0)
```

```python
print(model_fit.summary)
```

```python
residuals = pd.DataFrame(model_fit.resid)
residuals.plot()
plt.show()
residuals.plot(kind='kde')
plt.show()
print(residuals.describe())
```

```python
output=model_fit.forecast()[0]
```
